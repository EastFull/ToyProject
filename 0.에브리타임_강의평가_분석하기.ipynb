### 5.1. 강의평가 총평에는 강의의 기본적인 속성들이 많이 드러나 있을 것이다.

본 분석은 **전체 강의평가 총평에 많이 언급되는 명사들이 강의의 기본적인 속성**임을 밝히기 위해 진행되었다.<br>
분석의 수행은 2019학년도 1학기 소프트웨어융합대학에서 개설된 강의들에 대한 1025개의 강의평가를 이용하여 분석 결과를 도출한다.<br>
분석에 사용한 데이터는 **9. 별첨2** 의 **9.2.** 에 명시되어 있는 '1st_semester_lectures_evaluation.csv' 이다.

강의평가 총평 분석에 앞서, 강의의 기본적인 속성(과제, 시험, 학점 등)은 다른 강의도 가질 수 있는 강의 자체의 속성이라 정의한다.

분석 결과 도출을 위한 작업과 Python 코드는 아래에 명시한다.

#### 5.1.1. KoNLPy 형태소 분석을 통해 강의평가에서 명사들을 추출, list 생성

한국어로 이루어진 문장을 효과적으로 분석하기 위해 KoNLPy **(7. 참고문헌 명시)** 를 사용했다. <br>

```python
from konlpy.tag import Okt#형태소 분석
import csv#csv파일 입출력

def make_noun_list(input_list, index_of_text):
    """
    (list, int) -> list
   
    
    강의평가 list와 총평이 있는 index 를 입력으로 받아 문장의 명사만 추출하여 list로 값을 반환하는 함수
    """
    okt = Okt()
    result = []

    n = 1
    length = len(input_list)
    
    for content in input_list:
        print(n, "/", length, "번째 process...")
        #문장을 분할, 형태소 부착 
        tokenlist = okt.pos(content[index_of_text], stem=True, norm=True)
        temp=[]
    
        for word in tokenlist:
            #형태소가 명사일 때
            if word[1] in ["Noun"]:
                temp.append((word[0]))
        if temp:
            result.append(temp)
        n = n + 1
        
    return result

evaluation_list_for_basic = []
evaluation_noun_list_for_basic = []

with open('./1st_semester_processed_data/1st_semester_lectures_evaluation.csv', 'r', encoding = 'euc_kr') as evaluation_file:
    my_csv_reader = csv.reader(evaluation_file)
    for line in my_csv_reader:
        evaluation_list_for_basic.append(line)
        
evaluation_noun_list_for_basic = make_noun_list(evaluation_list_for_basic, 4)
```

#### 5.1.2. 명사 list를 기반으로 collections.Counter를 이용하여 wordcloud를 만들기 위한 dictionary list 생성

완성된 명사 list안에 존재하는 명사들의 출현 빈도를 집계하기 위해 collections.Counter **(7. 참고문헌 명시)** 를 사용한다.

```python
from collections import Counter
import random

r = lambda: random.randint(180,200)
color = lambda: (r(), r(), r())

def make_wordcloud_dict_list(noun_list, ntags = 30, multiplier = 1):
    """
    (list, int , int) -> list
   
    
    list를 입력으로 받아 list 안의 값들의 빈도수를 상위 ntags(default = 30)만큼 가져오고 multiplier로 wordcloud 안의 단어 크기 조절
    """
    count = Counter()
    return_list = []
    
    n = 1
    length = len(noun_list)
    
    for nouns in noun_list:
        print(n, "/", length, "번째 process...")
        temp_count = Counter(nouns)
        for key in temp_count:
            count.update({key : temp_count[key]})
        n = n + 1
        
    for word, word_count in count.most_common(ntags):
        #출현 빈도가 5 이상인 경우 진행
        if word_count >= 5 and len(word) > 1 :
            #wordcloud를 만들기위한 dicttionary list 만들기
            temp = {'color': color(), 'tag': word, 'size': word_count*multiplier}
            return_list.append(temp)
            
    return return_list

evaluation_dict_list = make_wordcloud_dict_list(evaluation_noun_list_for_basic)
print(evaluation_dict_list)
```

```python
import pytagcloud

copy_tags = []

for content in evaluation_dict_list:
    copy_tags.append(content.copy())

for content in copy_tags:
    content['size'] = content['size'] // 2

print(evaluation_dict_list)
print(copy_tags)

pytagcloud.create_tag_image(copy_tags, "./1st_semester_result_data/basic_wordcloud.png", fontname='Noto Sans CJK', size=(1300, 1300))
```

### 5.2. 강의평가는 전공 지식에 관련된 내용이 많지 않을 것이다.

본 분석은 **각 강의평가 총평에서 어떤 전공 지식을 쌓아갈 수 있는지 알기 어렵다**는 것을 밝히기 위해 진행되었다.<br>
분석의 수행은 2019학년도 1학기 소프트웨어융합대학에서 개설된 강의 중에 강의평가가 가장 많은 3개의 강의를 이용하여 분석 결과를 도출한다.  

분석에 사용한 데이터는 다음과 같다.<br>
* **9. 별첨2** 의 **9.1.** 에 명시되어 있는 '1st_semester_lectures.csv'
* **9. 별첨2** 의 **9.2.** 에 명시되어 있는 '1st_semester_lectures_evaluation.csv'

분석 결과 도출을 위한 작업과 Python 코드는 아래에 명시한다.

#### 5.2.1. 학수번호를 이용해 해당 강의의 강의평가가 위치한 index와 강의평가의 수를 얻기위한 함수 구현

```python
def search_with_code(code, evaluation_list):
    """
    (str, list) -> list
    
    string으로 이루어진 학수번호와 강의평가 list를 입력받아 해당 강의의 index와 강의평가의 수를 dictionary 형태로 알려주는 함수
    단, 강의평가 list가 학수번호로 정렬상태일 때 정상 동작
    """
    code_list = []
    
    for evaluation_info in evaluation_list:
        code_list.append(evaluation_info[0])
    
    return {'start' : code_list.index(code), 'length' : code_list.count(code)}
```

#### 5.2.2. 강의목록과 강의평가의 csv파일을 읽어와 각 list에 저장

```python
lecture_list_for_major = []
evaluation_list_for_major = []

with open('./1st_semester_processed_data/1st_semester_lectures.csv', 'r', encoding = 'euc_kr') as lecture_file:
    my_csv_reader = csv.reader(lecture_file)
    for line in my_csv_reader:
        lecture_list_for_major.append(line)

with open('./1st_semester_processed_data/1st_semester_lectures_evaluation.csv', 'r', encoding = 'euc_kr') as evaluation_file:
    my_csv_reader = csv.reader(evaluation_file)
    for line in my_csv_reader:
        evaluation_list_for_major.append(line)
        
#title 제거
lecture_list_for_major.remove(lecture_list_for_major[0])
evaluation_list_for_major.remove(evaluation_list_for_major[0])
```

#### 5.2.3. 강의목록의 학수번호를 따라 강의평가가 위치한 index와 강의평가의 수를 가지고 있는 dictionary list 생성

완성된 dictionary list는 강의평가 수가 많은 순으로 정렬한다.

```python
how_many_evaluation = []

for lecture in lecture_list_for_major:
    how_many_evaluation.append(search_with_code(lecture[1], evaluation_list_for_major))

how_many_evaluation = sorted(how_many_evaluation, key=lambda k: k['length'], reverse = True)
print(how_many_evaluation)  
```

#### 5.2.4. index와 강의평가의 수를 가지고 있는 dictionary list에서 상위 3개 강의에 대한 wordcloud 생성

index와 강의평가의 수를 가지고 있는 dictionary list에서 강의평가의 수가 많은 상위 3개의 강의평가를 추출한다.

```python
evaluation_ranking = []

for i in range(3):
    temp_list = []
    for j in range(how_many_evaluation[i]['start'], how_many_evaluation[i]['start'] + how_many_evaluation[i]['length']):
        temp_list.append(evaluation_list_for_major[j][:])
    evaluation_ranking.append(temp_list)

for i in range(3):
    print(evaluation_ranking[i])
```

추출된 상위 3개의 강의평가에서 **5.1.1.** 에 명시되어 있는 make_noun_list 함수를 이용, 각 강의평가에 대한 명사 list를 만든다.

```python
evaluation_ranking_noun_list = []

for evaluation_list in evaluation_ranking:
    evaluation_ranking_noun_list.append(make_noun_list(evaluation_list, 4))
    
for i in range(3):
    print(evaluation_ranking_noun_list[i])
```

## 10. 별첨3

### 10.1. HTML로 이루어진 강의목록의 데이터 원본을 가공하고 csv파일로 바꾸기 위한 소스코드

```python
import re #태그 제거를 위함
import csv #csv파일 입출력

def remove_tag(content):
    """
    (_io.TextIOWapper) -> str
    
    io.TextIOWrapper를 입력으로 받아 html 태그를 제거하고 string으로 값을 반환하는 함수
    """
    tag_cleaner = re.compile('<.*?>')
    cleantext = re.sub(tag_cleaner, ' ', content)
    return cleantext

def del_duplicate_space(content):
    """
    (str) -> list
    
    str을 입력으로 받아 str의 중복된 띄어쓰기를 없애고 띄어쓰기로 split하며 list를 반환하는 함수
    """
    return " ".join(content.split()).split()

def make_valid_lecture_data(content):
    """
    (list) -> list
   
    
    list를 입력으로 받아 list 안의 의미있는 값만 추출하여 list로 값을 반환하는 함수
    """
    valid_lecture_data = []
    
    i = 0
    while(i < len(content)): 
        if(content[i] == "조회"):
            #강의목록에서 가져오는 내용 : 학년, 학수 번호, 강의 명, 교수님 성함
            temp_list = [content[i + 1], content[i + 2], content[i + 3], content[i + 4]]
            valid_lecture_data.append(temp_list)
            i = i + 4;
        else:
            i = i + 1;
    
    return valid_lecture_data

def is_duplicate(checker_list, checkee_list):
    """
    (list, list) -> boolean
    
    두개의 강의정보 list를 입력 받아
    두 강의정보 list가 중복되는(강의 명과 교수님 성함이 같은)지 확인하고 그에 따른 boolean 값을 반환하는 함수
    """
    if((checker_list[2] == checkee_list[2]) and (checker_list[3] == checkee_list[3])):
        return True
    else:
        return False

def is_equal(checker_list, checkee_list):
    """
    (list, list) -> boolean
    
    두개의 강의정보 list를 입력 받아
    두 강의정보 list가 같은지 확인하고 그에 따른 boolean 값을 반환하는 함수
    """
    condition_0 = checker_list[0] == checkee_list[0]
    condition_1 = checker_list[1] == checkee_list[1]
    condition_2 = checker_list[2] == checkee_list[2]
    condition_3 = checker_list[3] == checkee_list[3]
    
    if(condition_0 and condition_1 and condition_2 and condition_3):
        return True
    else:
        return False
    
def remove_list(remove_list, entire_list):
    """
    (list, list) -> int
    
    지워야할 강의정보 list(학년, 학수 번호, 강의 명, 교수님 성함)들로 이루어진 list와 강의정보 list들로 이루어진 list를 입력 받아
    지워야할 강의정보 list를 강의정보 list들로 이루어진 list에서 지우는 함수
    """
    for lecture_info in remove_list:
        if(lecture_info in entire_list):
            entire_list.remove(lecture_info)
    
    return

cs_lecture_file = open('./1st_semester_original_data/1st_semester_cs.html', 'r')
#컴퓨터공학과 강의목록 만들기 : remove_tag -> del_duplicate_space -> make_valid_lecture_data
cs_data = make_valid_lecture_data(del_duplicate_space(remove_tag(cs_lecture_file.read())))
cs_lecture_file.close()

sw_lecture_file = open('./1st_semester_original_data/1st_semester_sw.html', 'r')
#소프트웨어융합학과 강의목록 만들기 : remove_tag -> del_duplicate_space -> make_valid_lecture_data
sw_data = make_valid_lecture_data(del_duplicate_space(remove_tag(sw_lecture_file.read())))
sw_lecture_file.close()

first_semester_lectures = cs_data + sw_data

#중복을 제거하기 전의 list
print("<Before Removing Dupliates> Length", len(first_semester_lectures))
for show_lecture in first_semester_lectures:
    print(show_lecture)
print("---------------------------------------------------------")

#동일한 강의 정보를 삭제(학년, 학수번호, 강의 명, 교수님 성함이 모두 일치하는 경우)
copy_lectures = first_semester_lectures[:]

for lecture_info in copy_lectures:
    if(first_semester_lectures.count(lecture_info) > 1): 
        first_semester_lectures.remove(lecture_info)
        
#강의 이름과 교수님의 성함이 같은(분반이 있는) 강의 정보를 삭제
copy_lectures = first_semester_lectures[:]
duplicate_list = []

for i in range(len(copy_lectures)):
    for j in range(i, len(copy_lectures)):
        #강의 이름과 교수님의 성함이 같은 강의(같은 강의가 아니라 분반인 강의)
        if(is_duplicate(copy_lectures[i], copy_lectures[j]) and not is_equal(copy_lectures[i], copy_lectures[j])):
            #분반이 3개 이상인 경우 같은 정보가 중복으로 들어갈 수 있으므로 이를 방지
            if(duplicate_list.count(copy_lectures[j]) == 0):
                duplicate_list.append(copy_lectures[j])
remove_list(duplicate_list, first_semester_lectures)

#완성된 list를 학수 번호를 기준으로 정렬
first_semester_lectures.sort(key=lambda lecture_info: lecture_info[1])

#최종 강의 list
print("<After Removing Dupliates> Length", len(first_semester_lectures))
for show_lecture in first_semester_lectures:
    print(show_lecture)
print("---------------------------------------------------------")

first_semester_lectures = [["학년", "학수번호", "강의 명", "교수님 성함"]] + first_semester_lectures

with open('./1st_semester_processed_data/1st_semester_lectures.csv', 'w', encoding = 'euc_kr', newline='') as writer_csv:
    myWriter = csv.writer(writer_csv)
    for i in range(len(first_semester_lectures)): 
        myWriter.writerow(first_semester_lectures[i])
```

### 10.2. 강의목록 csv 파일에서 가져온 학수번호에 해당되는 강의평가 데이터 원본을 가공하기 위한 소스코드

```python
def show_evaluation_list(evaluation_list):
    print("#####################################################################################")
    for evaluation in evaluation_list:
        print("학수 번호 :", evaluation[0])
        print("평균 별점 :", evaluation[1])
        print("-------------------------------------------------------------------------------------")
        for individual_evaluation in evaluation[2]: 
            print("별점 :", individual_evaluation[0])
            print("추천수 :", individual_evaluation[1])
            print("총평 :", individual_evaluation[2])
            print("-------------------------------------------------------------------------------------")
        print("#####################################################################################")
    return

lecture_list = []

#앞서 수행되었던 1학기 소프트웨어융합대학의 개설 강의 목록을 읽어온다.
with open('./1st_semester_processed_data/1st_semester_lectures.csv', 'r') as lectures_file:
    for lecture_info in lectures_file:
        lecture_list.append(lecture_info.strip('\n').split(','))

#태그제거
lecture_list.remove(lecture_list[0])

all_lectures_evaluation_list = []

#강의평가 html을 분석하여 필요한 정보를 가져옴
for lecture_info in lecture_list:
    file_name = './1st_semester_original_data/' + lecture_info[1] + '.html'
    with open(file_name, 'r') as lecture_evaluation_file:
        lecture_evaluation = lecture_evaluation_file.read()
        #강의평가의 평균 별점
        average_star_point = lecture_evaluation[lecture_evaluation.find('<span class="value">') + 20:lecture_evaluation.find('</span>')]
        average_star_point = float(average_star_point)
        
        individual_evaluation_list = []#[[별점0, 추천수0, 총평0],[별점1, 추천수1, 총평1], ...]
        individual_evaluation = []#[별점, 추천수, 총평]
        
        if(average_star_point != 0):
            #강의평가 시작, 끝 찾기
            article_start_index = lecture_evaluation.find('<article>')
            article_end_index = lecture_evaluation.find('</article>')
            while(article_start_index != -1):
                #강의평가의 개인 별점
                star_start_index = lecture_evaluation.find('<span class="on" style="width: ', article_start_index, article_end_index) + 31
                star_end_index = lecture_evaluation.find('%;"></span>', star_start_index, article_end_index)
                star_point = 5 * int(lecture_evaluation[star_start_index : star_end_index]) /100
                
                #강의평가의 추천수
                posvote_start_index = lecture_evaluation.find('<span class="posvote">', article_start_index, article_end_index)
                posvote_point = 0
                if(posvote_start_index != -1):
                    posvote_start_index = posvote_start_index + 22
                    posvote_end_index = lecture_evaluation.find('</span>', posvote_start_index, article_end_index)
                    posvote_point = int(lecture_evaluation[posvote_start_index:posvote_end_index])
                
                #강의평가 총평
                text_start_index = lecture_evaluation.find('<p class="text">', article_start_index, article_end_index) + 16
                text_end_index = lecture_evaluation.find('</p>', text_start_index, article_end_index)
                text_overall = lecture_evaluation[text_start_index : text_end_index]
                text_overall = text_overall.replace('<br>', ' ')
                
                individual_evaluation = [star_point, posvote_point, text_overall]
                individual_evaluation_list.append(individual_evaluation)
                
                article_start_index = lecture_evaluation.find('<article>', article_end_index + 10)
                article_end_index = lecture_evaluation.find('</article>', article_end_index + 10)
                
        all_lectures_evaluation_list.append([lecture_info[1], average_star_point, individual_evaluation_list])

show_evaluation_list(all_lectures_evaluation_list)

#가공한 데이터의 저장
with open('./1st_semester_processed_data/1st_semester_lectures_evaluation.csv', 'w', encoding = 'euc_kr', newline='') as writer_csv:
    myWriter = csv.writer(writer_csv)
    myWriter.writerow(['학수번호','평균별점','개인별점','추천수','총평'])
    for i in range(len(all_lectures_evaluation_list)): 
        if (len(all_lectures_evaluation_list[i][2]) == 0):
            myWriter.writerow([all_lectures_evaluation_list[i][0], 
                               all_lectures_evaluation_list[i][1], '#NODATA', '#NODATA', '#NODATA'])
        else:
            for j in range(len(all_lectures_evaluation_list[i][2])):
                myWriter.writerow([all_lectures_evaluation_list[i][0], 
                                   all_lectures_evaluation_list[i][1], 
                                   all_lectures_evaluation_list[i][2][j][0],
                                   all_lectures_evaluation_list[i][2][j][1],
                                   all_lectures_evaluation_list[i][2][j][2]])
```
